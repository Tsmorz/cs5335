Converting 2R-Manipulator joint angles to xy coordinates

```python
from matplotlib import pyplot as plt
import numpy as np
```

Circle in angle space to cartesian space

```python
def qcossin(Links,A):
    t = np.linspace(0,2*np.pi,500)
    q1 = A[0]*np.pi*np.cos(t)+A[1]*np.pi
    q2 = A[2]*np.pi*np.sin(t)+A[3]*np.pi

    l1 = Links[0]
    l2 = Links[1]
    x = l1*np.cos(q1) + l2*np.cos(q1+q2)
    y = l1*np.sin(q1) + l2*np.sin(q1+q2)

    plt.figure(1)
    plt.subplot(1,2,1)
    plt.plot(q1,q2)
    plt.subplot(1,2,2)
    plt.plot(x,y)
    plt.axis("equal")

    return

def qcos(Links,A):
    t = np.linspace(0,2*np.pi,500)
    q2 = A[0]*t
    q1 = A[1]*np.pi*np.cos(A[2]*t)+A[3]*np.pi

    l1 = Links[0]
    l2 = Links[1]
    x = l1*np.cos(q1) + l2*np.cos(q1+q2)
    y = l1*np.sin(q1) + l2*np.sin(q1+q2)

    plt.figure(1)
    plt.subplot(1,2,1)
    plt.plot(q1,q2)
    plt.subplot(1,2,2)
    plt.plot(x,y)
    plt.axis("equal")

    return

def qsin(Links,A):
    t = np.linspace(0,2*np.pi,500)
    q1 = A[0]*t
    q2 = A[1]*np.pi*np.sin(A[2]*t)+A[3]*np.pi

    l1 = Links[0]
    l2 = Links[1]
    x = l1*np.cos(q1) + l2*np.cos(q1+q2)
    y = l1*np.sin(q1) + l2*np.sin(q1+q2)

    plt.figure(1)
    plt.subplot(1,2,1)
    plt.plot(q1,q2)
    plt.subplot(1,2,2)
    plt.plot(x,y)
    plt.axis("equal")

    return
```

```python
Links = [1,1]
A = [1,1,2,1]
```

```python
qcossin(Links,A)
```

```python
qcos(Links,A)
```

```python
qsin(Links,A)
```
